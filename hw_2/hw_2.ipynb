Проверьте ваши векторайзеры на корпусе который использовали на вебинаре, составьте таблицу метод векторизации и скор который вы получили (в методах векторизации по изменяйте параметры что бы добиться лучшего скора) обратите внимание как падает/растёт скор при уменьшении количества фичей, и изменении параметров, так же попробуйте применить к векторайзерам PCA для сокращения размерности посмотрите на качество сделайте выводы

```python
# Загружаем данные
with open('corpus','r') as f:
    data = f.read()
labels, texts = [], []
for  line in data.split("\n"):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)
```

```python
def testing(vectorizer,model=LogisticRegression()):
    vectorizer.fit(trainDF['text'])

    xtrain =  vectorizer.transform(train_x)
    xvalid =  vectorizer.transform(valid_x)

    model.fit(xtrain, train_y)
    predictions = model.predict(xvalid)
    
    return accuracy_score(valid_y, predictions)
```

```python
# Проверим как работает
vec = TfidfVectorizer(max_df=0.9,max_features = 100,stop_words=stop_words)

testing(vec)
```

```python
max_df = [1,0.95,0.9,0.85,0.8,0.75,0.7]
min_df = [1,2,5,8,10,15,25,30,40,50]
max_features = [None,100,500,1000,2000,5000,10000,15000,20000,25000]
ngram_range=[(1, 1),(1,2),(2,2),(1,3),(2,3),(3,3)]
```

```python
def choice():
    ma_df = np.random.choice(max_df)
    mi_df = np.random.choice(min_df)
    m_features = np.random.choice(max_features)
    n_range = random.choice(ngram_range)
    key = str(ma_df)+'_'+str(mi_df) +"_" +str(m_features) + '_'+str(n_range)
    return key,ma_df,mi_df, m_features, n_range
    
```

```python
%%time
d={}
count = 1000
for _ in range(count):
    for i in range(5):
        key,ma_df,mi_df, m_features, n_range = choice()
        if key in d:
            continue
        else:
            break
    vec = TfidfVectorizer(max_df=ma_df,min_df =mi_df, max_features = m_features,ngram_range = n_range ,stop_words=stop_words) 
    d[key] = testing(vec) 
                
    
    
```

```python
%%time
d_2={}
count = 1000
for _ in range(count):
    for i in range(5):
        key,ma_df,mi_df, m_features, n_range = choice()
        if key in d_2:
            continue
        else:
            break
    vec = CountVectorizer(max_df=ma_df,min_df =mi_df, max_features = m_features,ngram_range = n_range ,stop_words=stop_words) 
    d_2[key] = testing(vec) 
                
    
```

```python
result = pd.DataFrame()
for  k,r in d.items():
    my_row = {'vectoreiz':'tfidf','param':k,'accuracy':r}
    result = result.append(my_row,ignore_index=True)
for  k,r in d_2.items():
    my_row = {'vectoreiz':'count','param':k,'accuracy':r}
    result = result.append(my_row,ignore_index=True)    
result.sort_values('accuracy',ascending = False,inplace = True)
```

Так же видно, что униформы не попали в топ лучших результатов, привалюруют (1,3) и (1,2).
